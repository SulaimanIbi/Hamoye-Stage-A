```python
def fuel_v_counts(fuel):
    df = data[data['fuel_type_code_pudl'] == fuel]
    return df['fuel_unit'].value_counts(dropna=False, normalize=True) * 100
```

```python
data_copy = data.copy()
```

```python
# rename the columns
columns = {'utility_id_ferc1':'utility_id', 'plant_name_ferc1':'plant_name', 'fuel_type_code_pudl':'fuel_type',
          'fuel_qty_burned':'qty_burned', 'fuel_mmbtu_per_unit':'mmbtu_per_unit', 'fuel_cost_per_unit_burned':'cost_per_unit_burned',
          'fuel_cost_per_unit_delivered':'cost_per_unit_delivered', 'fuel_cost_per_mmbtu':'cost_per_mmbtu'}

data_copy.rename(columns, inplace=True, axis=1)
```

```python
data_copy
```

```python
coal_conv_mmbtu = {'mmbtu':1, 'mcf':1.0279, 'ton':27.7782, 'bbl':5.799924}
gas_conv_mmbtu = {'mmbtu':1, 'mcf':1.0279, 'ton':39.6825, 'bbl':5.799924}
oil_conv_mmbtu = {'mmbtu':1, 'bbl':5.799924, 'gal':0.141, 'mcf':1.0279, 'ton':39.6825}
nuclear_conv_mmbtu = {'mmbtu':1, 'mwhth':3.412, 'mwdth':81.888, 'kgU':78479, 'gramsU':78.479}
waste_conv_mmbtu = oil_conv_mmbtu
other_conv_mmbtu = oil_conv_mmbtu
```

```python
# def coal_unit_converter(df):
#     df
#     return df
# unit_converter(data_copy)
```

```python
def unit_converter(df):
    # unit conversion for coal
    for unit in coal_conv_mmbtu:
        df.loc[(df['fuel_type'] == 'coal') & (df['fuel_unit'] == unit), 'qty_mmbtu_burned'] = coal_conv_mmbtu[unit] * df['qty_burned']
    
    # unit conversion for gas
    for unit in gas_conv_mmbtu:
        df.loc[(df['fuel_type'] == 'gas') & (df['fuel_unit'] == unit), 'qty_mmbtu_burned'] = gas_conv_mmbtu[unit] * df['qty_burned']
    
    # unit conversion for oil
    for unit in oil_conv_mmbtu:
        df.loc[(df['fuel_type'] == 'oil') & (df['fuel_unit'] == unit), 'qty_mmbtu_burned'] = oil_conv_mmbtu[unit] * df['qty_burned']
    
    # unit conversion for nuclear
    for unit in nuclear_conv_mmbtu:
        df.loc[(df['fuel_type'] == 'nuclear') & (df['fuel_unit'] == unit), 'qty_mmbtu_burned'] = nuclear_conv_mmbtu[unit] * df['qty_burned']
    
    # unit conversion for waste
    for unit in waste_conv_mmbtu:
        df.loc[(df['fuel_type'] == 'waste') & (df['fuel_unit'] == unit), 'qty_mmbtu_burned'] = waste_conv_mmbtu[unit] * df['qty_burned']
    
    # unit conversion for other
    for unit in other_conv_mmbtu:
        df.loc[(df['fuel_type'] == 'other') & (df['fuel_unit'] == unit), 'qty_mmbtu_burned'] = other_conv_mmbtu[unit] * df['qty_burned']
        
    return df

data_copy = unit_converter(data_copy)
```
